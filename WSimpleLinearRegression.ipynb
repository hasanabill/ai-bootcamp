```python
# Reshape the data to be column vectors
sqft_train = sqft_train.reshape(-1,1)
price_train = price_train.reshape(-1, 1)

print(sqft_train.shape)
print(price_train.shape)
```

```python
for s, p in zip(sqft_train[0:1000], price_train[0:1000]): #show first 10 elements for sqft and price together
  print(s,p)
```

```python
# Normalize the features (optional but recommended for gradient descent)
sqft_train_mean = np.mean(sqft_train)
sqft_train_std = np.std(sqft_train)
sqft_train_normalized = (sqft_train - sqft_train_mean) / sqft_train_std
```

```python
sqft_train_mean
```

```python
sqft_train_std
```

```python
sqft_train_normalized
```

```python
# Add a column of ones for the bias term
# it holds bias as the first column, and sqft_normaled (feature) at the second colum
X = np.hstack([np.ones_like(sqft_train_normalized), sqft_train_normalized])
#X = np.hstack((np.ones((sqft_normalized.shape[0], 1)), sqft_normalized)) #same as the upper one
print(X)
```

```python
X[0:10]
```

```python
X.shape
```

```python
# Initialize weights randomly
np.random.seed(42)
weights = np.random.randn(2, 1)
```

```python
weights
```

```python
weights.shape
```

```python
# Set hyperparameters
learning_rate = 0.01
num_iterations = 1000
```

```python
# Gradient Descent
for i in range(num_iterations):
    # Compute predictions
    predictions = np.dot(X, weights)

    # Compute error
    error = predictions - price_train

    # Compute gradients
    gradients = 1/len(price_train) * np.dot(X.T, error)

    # Update weights
    weights = weights - learning_rate * gradients

    # Denormalize weights
    intercept = weights[0][0] - weights[1][0] * sqft_train_mean / sqft_train_std
    slope = weights[1][0] / sqft_train_std

    if i%100==0:

        # Plot the data and regression line

        plt.figure(figsize=(8, 6))
        plt.scatter(sqft_train, price_train, color='blue', label='Data')
        #plt.plot(sqft_train, intercept + slope * sqft_normalized, color='red', label='Regression Line')
        plt.plot(sqft_train, intercept + slope * sqft_train, color='red', label='Regression Line')
        plt.title('Square Footage vs. Price')
        plt.xlabel('Square Footage')
        plt.ylabel('Price')
        plt.legend()
        plt.grid(True)
        plt.show()
        plt.pause(0.1)
```

```python
# Predict prices for some new square footages
new_sqft = np.array([[1500], [2000], [2500]])
#new_sqft_normalized = (new_sqft - sqft_mean) / sqft_std

# Calculate predicted prices using the learned model
predicted_prices = intercept + slope * new_sqft


print("Predicted prices:")
for i in range(len(new_sqft)):
    print(f"Square Footage: {new_sqft[i][0]}, Predicted Price: {predicted_prices[i][0]:.2f}")

# Denormalize weights
intercept = weights[0][0] - weights[1][0] * sqft_train_mean / sqft_train_std
slope = weights[1][0] / sqft_train_std

# # Print the equation of the regression line
print(f"Regression equation: Price = {intercept:.2f} + {slope:.2f} * sqft")

# # Plot the data and regression line
plt.figure(figsize=(8, 6))
plt.scatter(sqft_train, price_train, color='blue', label='Data')
#plt.plot(sqft_train, intercept + slope * sqft_normalized, color='red', label='Regression Line')
plt.plot(sqft_train, intercept + slope * sqft_train, color='red', label='Regression Line')
plt.title('Square Footage vs. Price')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()



```

```python
import numpy as np
```

```python
# Example usage:
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 1])  # OR Gate
```

```python
# Add bias term to input features
X_bias = np.hstack((np.ones((X.shape[0], 1)), X))

#X = np.hstack([np.ones_like(sqft_normalized), sqft_normalized])
```

```python
X_bias
```

```python
X_bias.shape
```

```python
# Initialize parameters
weights = np.zeros(X_bias.shape[1])
```

```python
weights
```

```python
learning_rate=0.01
num_iterations=10000
```

```python
def sigmoid(z):
        return 1 / (1 + np.exp(-z))
```

```python
X_bias.T
```

```python
y
```

```python
np.dot(X_bias.T,y)
```

```python
# Perform gradient descent
for _ in range(num_iterations):
    # Compute predictions
    linear_output = np.dot(X_bias, weights)
    y_pred = sigmoid(linear_output)

    # Compute gradient of binary cross-entropy loss w.r.t. parameters
    gradient = np.dot(X_bias.T, (y_pred - y)) / len(y)
    print(gradient)

    # Update parameters using gradient descent
    weights = weights - learning_rate * gradient
```

```python
weights
```

```python
X_test = np.array([[1, 1], [0, 0], [1, 0], [1, 1]])
X_test_bias = np.hstack((np.ones((X_test.shape[0], 1)), X_test))
```
